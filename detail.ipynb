```python
def get_detail(url):
    res = requests.get(url)
    soup = BeautifulSoup(res.text, 'html.parser')

    #job detail
    info = {}
    info['url'] = url
    try:
        job_detail = soup.find_all('div', {'class':'jobsearch-JobDescriptionSection-sectionItem'})
        for detail in job_detail:
            if detail.find('span'):
                info[detail.find_all('div')[0].text] = [i.text for i in detail.find_all('span')]
            else:
                info[detail.find_all('div')[0].text] = [i.text for i in detail.find_all('div')[1:]]
    except:
        pass
    
    #job qualification
    try:
        jobQual = soup.find('div', {'id':'qualificationsSection'})
        title = jobQual.find('h2').text
        info[title] = [i.text for i in jobQual.find_all('li', {'class':'icl-u-xs-p--none jobsearch-ReqAndQualSection-item'})]
    except:
        pass

    #job description
    try:
        jobdesc = soup.find('div', {'id':'jobDescriptionText'})
        info['Description'] = jobdesc.text
    except:
        print('No job')
        return None
        
    res.close()
    return info
```

```python
def details_loop(urls, start):
    details = []
    for i, url in enumerate(urls.tolist()):
        if i >= start:
            print(i)
            detail = get_detail(url)
            if detail == None:
                print('Get banned')
                break
            details.append(detail)
            wait = 5+randint(5, 15)
            sleep(wait)
    
    if len(details) > 0:
        df_detail = pd.DataFrame.from_dict(details)
        df_detail.to_csv(f'Data scientist details {start}.csv')
    return i, details
```

```python
# Salary analysis
df_Salary = df[~df['Salary'].isna()].reset_index()
df_Salary
```

```python
# full time job Salary
full_time = df_Salary.loc[df_Salary['Salary'].str.contains('a year')].reset_index()
list(full_time['Salary'])
```

```python
def min_salary(x):
    if re.search('- \$', x):
        return re.search('\$(.*?) -', x).group(1)
    else:
        return "0"

def max_salary(x):
    if (re.search('- \$', x)):
        return re.search('- \$(.*?) a', x).group(1)
    elif (re.search('\$(.*?) a', x)):
        return re.search('\$(.*?) a', x).group(1)
    else:
        return "0"
        
def salary(x):
    if (x['Min_Salary']==0):
        return x['Max_Salary']
    else:
        return x['Median_Salary']
```

```python
full_time['Min_Salary'] = full_time['Salary'].apply(min_salary)
full_time['Max_Salary'] = full_time['Salary'].apply(max_salary)
```

```python
full_time['Min_Salary'] = full_time['Min_Salary'].astype(str)
full_time['Max_Salary'] = full_time['Max_Salary'].astype(str)
full_time['Max_Salary'] = full_time['Max_Salary'].str.replace(',', '', regex=True).astype(float)
full_time['Min_Salary'] = full_time['Min_Salary'].str.replace(',', '', regex=True).astype(float)
full_time['Median_Salary'] = (full_time['Min_Salary'] + ((full_time['Max_Salary'] - full_time['Min_Salary'])/2))
full_time
```

```python
median_salary = round(full_time['Median_Salary'].sum()/full_time['Median_Salary'].count(), 2)
median_salary
```

```python
# qualification analysis
qulifications = list(df_detail[~df_detail['Qualifications'].isna()]['Qualifications'])
len(qulifications)
qulifications[0]
```

```python
pd.set_option("display.max_rows", None)
```

```python
allQ = [re.sub(f'["\'\[\]\)]', '',j) for i in qulifications for j in i.split(',')]
pd_skills = pd.DataFrame([i.split('(') for i in allQ], columns=['skills', 'required'])
pd_skills['required'].fillna(value="Required", inplace=True)
pd_skills['skills'] = pd_skills['skills'].str.strip()
pd_skills['skills'] = pd_skills['skills'].apply(lambda x: x.split(':')[0])
pd_cleaned = pd_skills[['skills', 'required']]
pd_cleaned['skills'] = pd_cleaned['skills'].str.strip()
pd_cleaned['skills'] = pd_cleaned['skills'].str.lower()
pd_cleaned['skills'] 
```

```python
#removeEdu
removeEdu = pd_cleaned.loc[~((pd_cleaned['skills'] == 'bachelors') | (pd_cleaned['skills'] == 'masters') | (pd_cleaned['skills'] == 'us work authorization') | (pd_cleaned['skills'] == 'high school or equivalent'))]
```

```python
required_skill = (removeEdu[removeEdu['required'] == 'Required']['skills'].value_counts().sort_values()/len(removeEdu[removeEdu['required'] == 'Required'])*100).apply(lambda x : round(x, 2))
required_skill
```

```python
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot()
p = ax.barh(required_skill.keys(), required_skill)
for i, rect in enumerate(required_skill.index):
    ax.text(required_skill[rect], i, required_skill[rect])
plt.title("Data scientist job required skills")
plt.xlabel('Percentage (%)')
```

```python
#preferred skill
preferred_skill = (removeEdu[removeEdu['required'] == 'Preferred']['skills'].value_counts().sort_values()/len(removeEdu[removeEdu['required'] == 'Preferred'])*100).apply(lambda x : round(x, 2))
```

```python
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot()
p = ax.barh(preferred_skill.keys(), preferred_skill)
for i, rect in enumerate(preferred_skill.index):
    ax.text(preferred_skill[rect], i, preferred_skill[rect])
plt.title("Data scientist job prefered skills")
plt.xlabel('Percentage (%)')
```
