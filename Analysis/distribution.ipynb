```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# data 
occupation = ['Data Scientist', 'Data Analyst', 'Data Architect', 'Data Engineer', 'Database Administrator', 'Statistician', 'Machine Learning Scientist', 'Machine Learning Engineer', 'Business Analyst']
amount = [365805, 521805, 882945, 236925, 65535, 30585, 267285, 660360, 562260]

df = pd.DataFrame(columns=['amount', 'occupation'])
df['amount'] = amount
df['occupation'] = occupation
explode = (0, 0, 0, 0, 0.5, 0, 0, 0, 0)
df = df.sort_values(by=['amount'])
df
```

```python

fig, axs = plt.subplots(figsize=[14, 6])
#bar char
axs.barh(df['occupation'], df['amount'])
for i, rect in enumerate(df['amount']):
    axs.annotate(rect, xy=(rect, i))
axs.set_title("Occupation distribution")
axs.set_xlabel('Occupations amounts')
#pie chart
# axs.axis('equal')
# axs.pie(df['amount'], labels=df['occupation'], explode=explode, autopct='%1.1f%%', shadow=True, radius=2)
# plt.show()
```
